```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df.isnull().sum()
```

Dropping Duplicates

```python
df_new=df.drop_duplicates()
```

```python
df_new.head()
```

Check the correlation of the columns

```python
plt.figure(figsize = (16, 10))
cor=df_new.corr()
sns.heatmap(cor, annot = True, cmap="YlGnBu")
plt.show()
```

Check the number of unique values for each column

```python
print(df_new.nunique())     
```

```python
#Creating a dataframe with top food name and ratings count

top_cuisine = df_new.groupby('Cuisines')['Food Rating'].sum().sort_values(ascending=False)
# printing top 5 country with highest ratings

for i,r in top_cuisine[:5].items():
    print(i,r)
```

Location wise Cuisines offered

```python
plt.figure(figsize=(15,8))

sns.countplot(df_new['Location'])
plt.ylabel("Count",fontsize=30)
plt.xlabel("Location",fontsize=30)
plt.show()
```

we can see most of the restaurants are from the location St. George,NY.

**List of Popular cuisine based on marital status and activity**

```python
plt.figure(figsize=(16, 14))

plt.subplot(2, 1, 1)
p = sns.barplot(x='Cuisines', y='Food Rating', hue='Marital Status', data=df_new, estimator=np.sum)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title('Popular cuisine based on marital status')
```

```python
plt.figure(figsize=(16, 14))
plt.subplot(2, 1, 2)
p = sns.barplot(x='Cuisines', y='Food Rating', hue='Activity', data=df_new, estimator=np.sum)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title('Popular cuisine based on activity')

plt.show()
```

**Popular cuisine based on Gender**

```python
plt.figure(figsize=(20,20))
q = sns.countplot(x='Cuisines', hue='Gender', data=df_new)
q.set_xticklabels(q.get_xticklabels(),rotation=30)
plt.title('Popular cuisine based on Gender')
plt.show()
```

we can see Japanese, Indian, and Filipino cuisines are preferred by male mostly, and French cuisines are preferred by female.

```python
plt.figure(figsize=(15,15))
sns.countplot(data=df_new, x='Cuisines', hue='Location', palette='mako')
plt.title('The Most Common Cuisines in each Locations')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**number of restaurants in each area**

```python
plt.figure(figsize=(10,10))

ax=df_new.Location.value_counts()
ax.plot(kind='pie',fontsize=20)

plt.title('number of restaurants in each area',fontsize=30,color='darkblue')
plt.show()
```
